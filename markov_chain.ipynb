```python
import random
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
file = open("oxxxymiron.txt", "r", encoding="utf8")
InputText = file.read()
# InputText = ""
PuncMarks = "!@#$%^&*()-+=_\"\',./\\№;:\n"
```

```python
def preprocess(text):
    for mark in PuncMarks:
        text = text.replace(mark, " ")
    return text.lower()
```

```python
TextToChain = f"{preprocess(InputText)}"
```

```python
vectorizer = CountVectorizer(ngram_range=(2,2), stop_words=[])
vectorizer.fit([TextToChain])
pairs = [f"*START* {TextToChain.split()[0]}"] + vectorizer.get_feature_names() + [f"{TextToChain.split()[-1]} *END*"]
```

```python
chainsRaw = [pair.split() for pair in pairs]
```

```python
chains = {}
for chainRaw in chainsRaw:
    if chainRaw[0] in chains:
        if not chainRaw[1] in chains[chainRaw[0]]:
            chains[chainRaw[0]].append(chainRaw[1])
    else:
        chains[chainRaw[0]] = []
        chains[chainRaw[0]].append(chainRaw[1])
```

```python
text = ""
currentWord = "*START*"
while True:
    currentWord = chains[currentWord][random.randint(0, len(chains[currentWord]) - 1)]
    if currentWord == "*END*":
        text = text[:-1]
        break
    text += currentWord + " "
text
```
